## Importacion de librerias relevantes


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
from statsmodels.tsa.arima.model import ARIMA
from scipy.stats.distributions import chi2
import statsmodels.tsa.stattools as sts 
import seaborn as sns
sns.set()
```

## Importacion y preprocesamiento de datos

```python
raw_csv_data = pd.read_csv("Index2018.csv") 
df_comp=raw_csv_data.copy()
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = True)
df_comp.set_index("date", inplace=True)
df_comp=df_comp.asfreq('b')
df_comp=df_comp.fillna(method='ffill')
```

```python
df_comp['market_value']=df_comp.ftse
```

```python
del df_comp['spx']
del df_comp['dax']
del df_comp['ftse']
del df_comp['nikkei']
size = int(len(df_comp)*0.8)
df, df_test = df_comp.iloc[:size], df_comp.iloc[size:]
```

## LLR Test

```python
def LLR_test(mod_1, mod_2, DF=1):
    L1 = mod_1.fit().llf
    L2 = mod_2.fit().llf
    LR = (2*(L2-L1))
    p = chi2.sf(LR, DF).round(3)
    return p
```

## The DF-Test

```python
sts.adfuller(df.market_value)
```

## Using Returns

El uso de los retornos de las acciones es una práctica común en el análisis financiero y en la modelización en finanzas. Los retornos de las acciones se calculan para evaluar el rendimiento de una inversión en acciones a lo largo del tiempo y son esenciales para una variedad de aplicaciones, como la valoración de activos, la toma de decisiones de inversión y la gestión de riesgos. 
$$\frac{P_{t}-P_{t-1}}{P_{t-1}}$$

```python
df['returns'] = df.market_value.pct_change(1).mul(100)
df = df.iloc[1:]
```

```python
sts.adfuller(df.returns)
```

## ACF and PACF for Returns

Grafico de autocorrelacion parcial y completa de los retornos del activo financiero **FTSE**

```python
sgt.plot_acf(df.returns, lags=40, zero = False)
plt.title("ACF FTSE Returns", size=24)
plt.show()
```

```python
sgt.plot_pacf(df.returns, lags = 40, zero = False, method = ('ols'))
plt.title("PACF FTSE Returns", size=24)
plt.show()
```

## AR(1) para retornos

```python
model_ret_ar_1 = ARIMA(df.returns, order = (1,0,0))
```

```python
results_ret_ar_1 = model_ret_ar_1.fit()
```

```python
results_ret_ar_1.summary()
```

## Modelos AR con mayores rezagos de los retornos

```python
model_ret_ar_2 = ARIMA(df.returns, order = (2,0,0))
results_ret_ar_2 = model_ret_ar_2.fit()
results_ret_ar_2.summary()
```

```python
LLR_test(model_ret_ar_1, model_ret_ar_2)
```

```python
model_ret_ar_3 = ARIMA(df.returns, order = (3,0,0))
results_ret_ar_3 = model_ret_ar_3.fit()
results_ret_ar_3.summary()
```

```python
LLR_test(model_ret_ar_2, model_ret_ar_3)
```

## Normalizando valores

Prueba de normalidad de los datos

```python
benchmark = df.market_value.iloc[0]
```

```python
df['norm'] = df.market_value.div(benchmark).mul(100)
```

```python
sts.adfuller(df.norm)
```

```python
##Prueba de esctacionariedad
bench_ret = df.returns.iloc[0]
df['norm_ret'] = df.returns.div(bench_ret).mul(100)
sts.adfuller(df.norm_ret)
```

## Retornos normalizados

```python
model_norm_ret_ar_2 = ARIMA(df.norm_ret, order=(2,0,0))
results_norm_ret_ar_2 = model_norm_ret_ar_2.fit()
results_norm_ret_ar_2.summary()
```
